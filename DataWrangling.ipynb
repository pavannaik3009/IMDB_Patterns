```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import re
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
df = pd.read_csv('movie_list.csv')
```

```python
df.head()
```

```python
# indexing the dataframe
df.index = list(range(1,251))
```

The release date has country listed taht needs to be removed.

```python
df['Country'].unique()
```

```python
#removing the country name from the Release Date column
df['Release Date'] = df['Release Date'].apply(lambda x: x.split('(')[0] )
```

```python
# Add Release year(extracting only the year from the date column) and drop release date
yearregex = re.compile('\d\d\d\d')
def func(x):
    result = yearregex.search(x)
    return(result.group())
df['Release Year'] = df['Release Date'].apply(func)
df.drop(['Release Date'], axis=1, inplace=True)
```

```python
df.head(10)
```

'See More' is not a language in the language column. The language is present in the country column.

```python
langmiss = df.loc[df['Language']=='See more'].index
```

```python
langmiss
```

```python
for i in langmiss:
    df.ix[i,4] = df.ix[i,3]
```

```python
df.Country.unique()
```

```python
langcorr = {'Japanese':'Japan','Italian':'Italy','English':'USA','Hindi':'India','German':'Germany',
           'French':'France', 'Swedish':'Sweden', 'Belarusian':'Soviet Union', 'Russian':'Russia', 
           'Turkish':'Turkey', 'Korean':'South Korea'}
```

```python
for i in range(1,251):
    if df['Country'].ix[i] in langcorr.keys():
        df['Country'].ix[i] = langcorr[df['Country'].ix[i]]
```

```python
df.head(10)
```

The IMDB rating is a float value and not a string value. It needs to be converted

```python
df['IMDB Rating'] = df['IMDB Rating'].apply(lambda x: float(x))
```

The runtime data comprises of hours and minutes. Converting it to only minutes.

```python
def func(x):
    a = int(x.split('h')[0])*60 
    b = x.split('h')[1].split('min')[0].strip()
    if b=='':
        b=0
    else:
        b = int(x.split('h')[1].split('min')[0].strip())
    return(a+b)
df['Running Time'] = df['Running Time'].apply(func)
```
